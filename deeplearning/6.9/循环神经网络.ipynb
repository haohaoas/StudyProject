```python
import numpy as np
import torch.nn as nn
import torch
#创建数据集
def create_dataset():
    sentence = (
    "It was the best of times, it was the worst of times, "
    "it was the age of wisdom, it was the age of foolishness, "
    "it was the epoch of belief, it was the epoch of incredulity, "
    "it was the season of Light, it was the season of Darkness, "
    "it was the spring of hope, it was the winter of despair, "
    "we had everything before us, we had nothing before us, "
    "we were all going direct to Heaven, we were all going direct the other way – "
    "in short, the period was so far like the present period, that some of its noisiest authorities "
    "insisted on its being received, for good or for evil, in the superlative degree of comparison only. "
    "There were a king with a large jaw and a queen with a plain face, on the throne of England; "
    "there were a king with a large jaw and a queen with a fair face, on the throne of France. "
    "In both countries it was clearer than crystal to the lords of the State preserves of loaves and fishes, "
)
    sentence=sentence.lower()
    #创建字符集
    chars_set=list(set(sentence))
    char_dict={ch: i  for i,ch in enumerate(chars_set)}
    idx_dict={i: ch for i, ch in enumerate(chars_set)}
    n_chars=len(chars_set)
    #构造训练数据
    x_data=[char_dict[c] for c in sentence[:-1]]
    y_data=[char_dict[c] for c in sentence[1:]]
    def one_hot(idx,depth):
        vec=torch.zeros(len(idx),depth)
        for i,val in enumerate(idx):
            vec[i,val]=1
        return vec
    x=one_hot(x_data,n_chars).unsqueeze(0)
    y=torch.tensor(y_data).unsqueeze(0)
    return x,y,char_dict,idx_dict,n_chars
```

```python
#定义模型
class CharModel(nn.Module):
    def __init__(self,mode='rnn',input_size=0,hidden_size=128,num_classes=0):
        super(CharModel,self).__init__()
        if mode=='rnn':
            self.rnn=nn.RNN(input_size,hidden_size,batch_first=True)
        elif mode=='gru':
            self.rnn=nn.GRU(input_size,hidden_size,batch_first=True)
        elif mode=='lstm':
            self.rnn=nn.LSTM(input_size,hidden_size,batch_first=True)
        self.fc=nn.Linear(hidden_size,num_classes)
    def forward(self,x):
        out,h=self.rnn(x)
        out=self.fc(out)
        return out
```

```python
#训练
def train(model,name):
    criterion=nn.CrossEntropyLoss()
    optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
    for epoch in range(300):
        optimizer.zero_grad()
        output=model(x)
        print(output.shape)
        print(y.shape)
        loss=criterion(output.view(-1,output.size(-1)),y.view(-1))
        loss.backward()
        optimizer.step()
        if (epoch + 1) % 50 == 0:
            pred = output.argmax(dim=2).squeeze().tolist()
            pred_str = ''.join([idx_dict[i] for i in pred])
            print(f"[{name}] Epoch {epoch+1}, Loss: {loss.item():.4f}, Predicted: {pred_str}")

```

```python
if __name__ == '__main__':
    x,y,char_dict,idx_dict,n_chars=create_dataset()
    model=CharModel(mode='rnn',input_size=n_chars,hidden_size=128,num_classes=n_chars)
    train(model,'rnn')
    model=CharModel(mode='lstm',input_size=n_chars,hidden_size=128,num_classes=n_chars)
    train(model,'lstm')
```
